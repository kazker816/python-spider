```python
from lxml import etree
text = """
<div>
    <ul>
         <li class="item-0"><a href="link1.html">first item</a></li>
         <li class="item-1"><a href="link2.html">second item</a></li>
         <li class="item-inactive"><a href="link3.html">third item</a></li>
         <li class="item-1"><a href="link4.html">第四项</a></li>
         <li class="item-0"><a href="link5.html">第五项</a>
     </ul>
 </div>
"""
#利用etree.HTML，将字符串解析为HTML文档
httpElement = etree.HTML(text)
print(type(httpElement))  #<class 'lxml.etree._Element'>
# 按字符串序列化HTML文档
print(etree.tostring(httpElement,encoding='utf-8').decode('utf-8'))
```

```python
from lxml import etree
parser = etree.HTMLParser(encoding = 'utf-8')  
#使用html的解析器来解析Html文件，默认是xml的解析器哦
htmlElement = etree.parse('result.html',parser= parser)
print(etree.tostring(htmlElement,encoding='utf-8',pretty_print=True).decode('utf-8'))
# 输入结果和之前是相同的
```
