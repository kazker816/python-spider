```python
import requests
import bs4
from bs4 import BeautifulSoup
def getHTMLtext(url,hd):
    try:
        r = requests.get(url,headers = hd,timeout = 10)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return ''

def htmlTOlist(univlist,html):
    soup = BeautifulSoup(html,'html.parser')
    #for every_univ in soup.find('tbody').children:
    for every_univ in soup.find_all('tr',attrs={"class":"alt"}):
        if isinstance(every_univ,bs4.element.Tag):
            tds = every_univ.find_all('td')
            univlist.append([tds[0].string,tds[1].string,tds[2].string,tds[3].string])

def printUNIVlist(univlist,num):
    tplt = "{0:^5}\t{1:{3}^10}\t{2:^10}\t{4:^5}"
    print(tplt.format("排名", "学校名称", "省份", chr(12288),'总分'))
    for i in range(num):  # 后来做了优化，是输出的学校名称居中对齐，chr（12288）是汉字里面的空格
        u = univlist[i]
        print(tplt.format(u[0], u[1], u[2], chr(12288),u[3]))

def main():
    url = 'http://www.zuihaodaxue.com/zuihaodaxuepaiming2019.html'
    resultlist = []
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'}
    htmltext = getHTMLtext(url,header)
    if htmltext=='':
        print('获取网站数据出错')
        return None
    htmlTOlist(resultlist,htmltext)
    printUNIVlist(resultlist,20)
    # print(len(resultlist))

main()
```

```python
import requests
import bs4
import random
from bs4 import BeautifulSoup
def getHTMLtext(url,hd):
    try:
        r = requests.get(url,headers = hd,timeout = 10)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text
    except:
        return ''

def HtmlTolist(resultlist,htmltext):
    item_name_list = []
    item_brief_introduction_list = []
    item_Province_list = []
    item_Type_list = []
    item_Total_Money_list = []
    item_Support_Num_list = []
    item_Finish_Rate_list = []
    soup = BeautifulSoup(htmltext,'html.parser')
    for every_item in soup.find_all('a',attrs={"class":"siteCardICH3"}):
        if isinstance(every_item,bs4.element.Tag):
            item_name_list.append(every_item.string)
    for every_item in soup.find_all('p',attrs={"class":"siteCardIC_p souSuo"}):
        if isinstance(every_item,bs4.element.Tag):
            item_brief_introduction_list.append(every_item.string)
    for every_item in soup.find_all('div',attrs={"class":"siteCardFLabelBox siteIlB_box"}):
        if isinstance(every_item,bs4.element.Tag):
            a_list = every_item.find_all('a')
            item_Province_list.append(a_list[0].string)
            try:
                item_Type_list.append(a_list[2].string)
            except:
                item_Type_list.append(a_list[1].string)
    for every_item in soup.find_all('div',attrs={"class":"siteCardFData"}):
        if isinstance(every_item,bs4.element.Tag):
            div_list = every_item.find_all('div')
            item_Total_Money_list.append(int(div_list[0].p.string[1:]))
            item_Support_Num_list.append(int(div_list[1].p.string))
            item_Finish_Rate_list.append(div_list[2].p.string)
    item_Inf_list = [[a,b,c,d,e,f] for a,b,c,d,e,f in zip(item_brief_introduction_list,item_Province_list,item_Type_list,item_Total_Money_list,item_Support_Num_list,item_Finish_Rate_list)]
    num_item = len(item_name_list)
    for i in range(num_item):
        resultlist.append({item_name_list[i]:item_Inf_list[i]})

def print_result(resultlist,num = 0):
    if num == 0:
        num =len(resultlist)
    for i in range(num):
        print(resultlist[i])

def main():
    url = 'http://www.zhongchou.cn/browse/id-28-re'
    resultlist = []
    header_list = [{ 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'},
                   { 'User-Agent':"Mozilla/5.0 (Macintosh; U; Mac OS X Mach-O; en-US; rv:2.0a) Gecko/20040614 Firefox/3.0.0 "},
                   { 'User-Agent':"Mozilla/5.0 (Macintosh; U; PPC Mac OS X 10.5; en-US; rv:1.9.0.3) Gecko/2008092414 Firefox/3.0.3"},
                   { 'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1"},
                   { 'User-Agent':"Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57 Safari/536.11"},
                   { 'User-Agent':"Opera/9.80 (Windows NT 6.1; U; en) Presto/2.8.131 Version/11.11"},
                   { 'User-Agent':"Opera/9.80 (Android 2.3.4; Linux; Opera mobi/adr-1107051709; U; zh-cn) Presto/2.8.149 Version/11.10"}]
    header = random.choice(header_list)
    htmltext = getHTMLtext(url,header)
    if htmltext=='':
        print('获取网站数据出错')
        return None
    HtmlTolist(resultlist,htmltext)
    print_result(resultlist)
    print(len(resultlist))
main()
```

```python
import requests
from bs4 import BeautifulSoup

DATA = []
def getHTMLtext(url,headers,timeout=5):
    try:
        response = requests.get(url,headers= headers,timeout = timeout)
        response.raise_for_status
        response.encoding = 'utf-8'
        return response.text
    except:
        return ''
    
def bs4Parser(text):
    soup =BeautifulSoup(text,'html5lib')   #html5lib
    day0 = soup.select_one('body > div.lqcontentBoxH > div.contentboxTab > div > div > div.hanml > div:nth-child(1)')
    province_table = day0.select('body > div.lqcontentBoxH > div.contentboxTab > div > div > div.hanml > div:nth-child(1) > div > table')
    for every_province in province_table:
        trs = every_province.select('table > tbody > tr')[2:]  #前两个是无用的哦
        for index,tr in enumerate(trs):
            tds = tr.select('tr> td')
            if index == 0:
                city = tds[1].get_text().strip()  #这里是为了去除提取的字符串前后的的空格，string提取到的是\n哦，所以要小心两者的区别
            else:
                city = tds[0].get_text().strip()
            min_temp = int(tds[-2].string)
            data = {'city':city,'min_temp':min_temp}
            DATA.append(data)
     

# 笔记：
# 这里用的全部是从浏览器里面直接复制出来的css选择器代码
# 如果使用的是lxml解析器，会发现不能得到想要的结果 
# 这是由于这个代码不规范引起的，我们可以使用html5lib解析器  pip install html5lib
# 这样会使得解析的速度减低，但是容错率上升了
        
if __name__ == '__main__':
    url = 'http://www.weather.com.cn/textFC/db.shtml'
    headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'
    }
    text = getHTMLtext(url,headers)
    if text != '':
        bs4Parser(text)
        DATA.sort(key = lambda x: x['min_temp'])
        top10 = DATA[:10]
        top10city = [i['city'] for i in top10]
        top10temp = [i['min_temp'] for i in top10]
        for i in range(10):
            print('{}:{}'.format(top10city[i],top10temp[i]))
        
    else:
        print('页面数据获取失败')
```

```python
import re
import csv
with open('语文单选.txt','r',encoding= 'gbk') as f:
    str_list = f.readlines()
    strs = ''.join(str_list)
r_question = re.compile(r'(\d{1,2}、.*?\s*?[ABCD]\s*?)')
questions = re.findall(r_question,strs)
r_choose = re.compile(r'(A[、．\.].*?C[、．\.].*?\n\s*(D[、．\.].*?\n)?)',re.DOTALL)
row_chooses =  re.findall(r_choose,strs)
chooses = []
for every_one in row_chooses:
    chooses.append(every_one[0].replace('\n','  '))
with open('语文.csv','w',encoding = 'utf-8',newline='') as f:
    writer = csv.writer(f)
    for every_one,choose in zip(questions,chooses):
        writer.writerow([(re.split('\(|（',every_one)[0].strip()),(re.split('\(|（',every_one)[1].strip())])
        writer.writerow([(choose),''])
```

```python
import threading
import time
def writing():
    for i in range(3):
        print('第{}次写作业'.format(i+1))
        time.sleep(0.5)
def drawing():
    for i in range(3):
        print('第{}次画画'.format(i+1))
        time.sleep(0.5)
def single_thread():
    writing()
    drawing()
def multi_thread():
    t1 = threading.Thread(target=writing)
    t2 = threading.Thread(target=drawing)
    t1.start()
    t2.start()
    

if __name__ =='__main__':
    print(threading.current_thread())
    multi_thread()
    
    
```
